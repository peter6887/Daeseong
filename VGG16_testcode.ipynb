```python
from keras.models import Model
from keras.preprocessing import image
from keras.optimizers.legacy import SGD
from keras.applications.vgg16 import VGG16, decode_predictions
import matplotlib.pyplot as plt
import numpy as np
import cv2
from pylab import imread
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
import io

def display_image(image):
  fig = plt.figure(figsize=(20, 15))
  plt.grid(False)
  plt.imshow(image)
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
image_data = uploaded[filename]

image = Image.open(io.BytesIO(image_data))
imagedata = np.array(image)

display_image(imagedata)
```

```python
model = VGG16(weights = 'imagenet', include_top=True)
sgd = SGD(learning_rate=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd, loss='categorical_crossentropy')

image = np.array(imagedata)

im = cv2.resize(image, (224,224))
im = np.expand_dims(im, axis = 0)

out = model.predict(im)
decode_predic = decode_predictions(out, top=3)[0]

mostsim, secsim, thdsim = decode_predic

display_image(im[0])
print('1. %s (%.2f%%)'% (mostsim[1], mostsim[2]*100))
print('2. %s (%.2f%%)'% (secsim[1], secsim[2]*100))
print('3. %s (%.2f%%)'% (thdsim[1], thdsim[2]*100))
```
